```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
```

```python
data = pd.read_csv('learningSet.csv')
```

```python
#creating a categorical df:
categorical = data.select_dtypes([object])
categorical.head()
```

```python
#checking for NUll values in all the columns:
categorical.isna().sum()
```

```python
#It looks like there are no Nan values, but there are a lot of empty cells that is an equivalent. 
#We will replace empty spaces with Nan
```

```python
categorical["MAILCODE"].value_counts()
```

```python
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
#checking for NUll values again and we are getting anotehr picture:
round(categorical.isna().sum()/len(categorical),4)*100
```

```python
#creating a new empty list and assigning 2 columns there:
drop_list = [cols for cols in categorical if cols in ["OSOURCE", "ZIP"]]
drop_list
```

```python
#identifying columns with more than 85% of missing values and adding them to drop_list:
def get_null_columns(df, threshold=0.85):
    null_percentages = df.isnull().sum() / len(df)
    return list(null_percentages[null_percentages > threshold].index)
```

```python
drop_list.extend(get_null_columns(categorical))
print(drop_list)
```

```python
#removing the columns included in a drop_list:
categorical_new = categorical.drop(columns=drop_list)
```

```python
print(categorical.shape)
print(categorical_new.shape)
```

```python
#working with Gender
categorical_new['GENDER'].value_counts()
```

```python
categorical_new['GENDER'].unique()
```

```python
categorical_new['GENDER'] = categorical_new['GENDER'].replace(" ", 'F')
```

```python
categorical_new['GENDER'].value_counts()
```

```python
df = pd.DataFrame(categorical_new["GENDER"].value_counts()).reset_index()

df.columns =["gender", "count"]
other = list(df[df["count"]< 2000]["gender"])

def clean_gender(x):
    if x in other:
        return "other"
    else:
        return x

categorical_new["GENDER"] = list(map(clean_gender, categorical_new["GENDER"]))
```
